```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
EPOCHS = 10

data = pd.read_table("./data/EURUSD1440.csv", names=["date", "time", "open", "high", "low", "close", "volume"])

num = data.shape[0]
```
