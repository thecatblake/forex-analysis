```python
import requests
from bs4 import BeautifulSoup
import xml.dom.minidom
```

```python
def getPage(url):
    req = requests.get(url)
    return BeautifulSoup(req.text, 'html.parser')
```

```python
url = "https://www.ft.com/companies"
parser = getPage(url)
```

```python
for a in parser.find_all("a"):
    if "/content/" in a["href"]:
        print(a.text, a["data-trackable"])
```
